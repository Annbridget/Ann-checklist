```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Airbnb_Open_Data.csv')

# Display the first few rows
df.head()
```

```python
# Check dataset info (columns, data types, missing values)
df.info()

# Check for missing values
df.isnull().sum()

# Summary statistics for numerical columns
df.describe()
```

```python
# Count missing values per column
df.isnull().sum()
```

```python
df.drop(['license', 'house_rules'], axis=1, inplace=True)
```

```python
# Remove dollar signs and commas, then convert to float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df['price'].fillna(df['price'].median(), inplace=True)
df['service fee'].fillna(df['service fee'].median(), inplace=True)
df['minimum nights'].fillna(df['minimum nights'].median(), inplace=True)
df['availability 365'].fillna(df['availability 365'].median(), inplace=True)
```

```python
print(df.columns)
```

```python
df['cancellation_policy'].fillna(df['cancellation_policy'].mode()[0], inplace=True)
df['room type'].fillna(df['room type'].mode()[0], inplace=True)
```

```python
df['last review'].fillna("Unknown", inplace=True)
```

```python
df['reviews per month'].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
```

```python
df['number of reviews'].fillna(0, inplace=True)
df['review rate number'].fillna(0, inplace=True)
df['calculated host listings count'].fillna(0, inplace=True)

```

```python
df['host name'].fillna("Unknown", inplace=True)
df['host_identity_verified'].fillna("Unknown", inplace=True)
df['host id'].fillna(0,inplace=True)
```

```python
df['NAME'].fillna("Unknown",inplace=True)
```

```python
df['instant_bookable'].fillna("unknown",inplace = True)
df['country code'].fillna(0, inplace=True)
df['country'].fillna("unknown",inplace = True)
df['long'].fillna(0, inplace=True)
```

```python

df['lat'].fillna(0, inplace=True)
df['neighbourhood'].fillna("unknown", inplace=True)
df['neighbourhood group'].fillna("unknown", inplace=True)

```

```python
df.isnull().sum()
```

```python
df['Construction year'].fillna(0,inplace =True)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Distribution of prices
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
#Room type count
plt.figure(figsize=(6,4))
sns.countplot(y=df['room type'], order=df['room type'].value_counts().index)
plt.title('Room Type Count')
plt.xlabel('Count')
plt.ylabel('Room Type')
plt.show()
```

```python
# price vs room type
plt.figure(figsize=(8,5))
sns.boxplot(x=df['room type'], y=df['price'])
plt.title('Price Variation by Room Type')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 most common neighbourhoods
df['neighbourhood'].value_counts().head(10).plot(kind='barh', figsize=(10,5), color='skyblue')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.title('Top 10 Neighbourhoods with Most Listings')
plt.show()
```

```python
#Average price by room type
df.groupby('room_type')['price'].mean().plot(kind='bar', color=['blue', 'orange', 'green'])
plt.ylabel('Average Price')
plt.title('Average Price per Room Type')
plt.show()
```
